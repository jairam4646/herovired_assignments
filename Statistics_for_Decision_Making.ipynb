```python
#1
# For the suburb of Altona, it is postulated that a typical property sells for $800,000
# • Use the data at hand to test this assumption.
# • Is the typical property price really $800,000 or has it increased?
# • Use a significance level of 5%.

altona_data = df[df['Suburb'] == 'Altona']
prices = altona_data['Price'].dropna()
mu = 800000
t_stat, p_value = stats.ttest_1samp(prices, mu)
sample_mean = prices.mean()
alpha = 0.05
print("Sample Mean Price in Altona:", sample_mean)
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < alpha:
    print("Reject the Null Hypothesis")
    if sample_mean > mu:
        print("The typical property price in Altona has increased above $800,000.")
    else:
        print("The typical property price in Altona is less than $800,000.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("The typical property price in Altona is approximately $800,000.")
```

```python
#2
# For the year 2016, is there any difference in the prices of properties sold in the summer months vs winter months?
# • Consider months from October till March as winter months and rest as summer months.
# • Use a significance level of 5%

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df_2016 = df[df['Date'].dt.year == 2016]
winter_months = [10, 11, 12, 1, 2, 3]
winter_prices = df_2016[df_2016['Date'].dt.month.isin(winter_months)]['Price'].dropna()
summer_prices = df_2016[~df_2016['Date'].dt.month.isin(winter_months)]['Price'].dropna()
t_stat, p_value = stats.ttest_ind(winter_prices, summer_prices, equal_var=False)
alpha = 0.05
print("Winter Mean Price:", winter_prices.mean())
print("Summer Mean Price:", summer_prices.mean())
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < alpha:
    print("Reject the Null Hypothesis: Prices are different between summer and winter.")
else:
    print("Fail to Reject the Null Hypothesis: No significant difference in prices.")
```

```python
# 3
# For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have a car parking space?
# • Use the column car in the dataset.
# • Round off your answer to 3 decimal places.

from scipy.stats import binom
abbotsford = df[df['Suburb'] == 'Abbotsford']
p = (abbotsford['Car'] == 0).sum() / len(abbotsford)
probability = binom.pmf(3, 10, p)
print("Probability:", round(probability, 3))
```

```python
# 4
# In the suburb of Abbotsford, what are the chances of finding a property with 3 rooms? Round your answer to 3 decimal places.

abbotsford = df[df['Suburb'] == 'Abbotsford']
probability = (abbotsford['Rooms'] == 3).sum() / len(abbotsford)
print("Probability:", round(probability, 3))
```

```python
# 5
# In the suburb of Abbotsford, what are the chances of finding a property with 2 bathrooms? Round your answer to 3 decimal places.

probability = (abbotsford['Bathroom'] == 2).sum() / len(abbotsford)
print("Probability:", round(probability, 3))
```

```python
# 6
# One-Sample Hypothesis Test (Industry Pricing) A real estate firm claims that the average property price in Richmond is $1,000,000.
# Using the dataset, test whether the actual average price is significantly different from this claim at a 5% significance
# level. Clearly state:
# • Null and alternative hypotheses
# • Test statistic
# • p-value
# • Final business conclusion


richmond = df[df['Suburb'] == 'Richmond']
prices = richmond['Price'].dropna()
mu = 1000000
t_stat, p_value = stats.ttest_1samp(prices, mu)
alpha = 0.05
print("Null Hypothesis (H0): The average property price in Richmond = $1,000,000")
print("Alternative Hypothesis (H1): The average property price in Richmond ≠ $1,000,000")
print("\nSample Mean Price:", prices.mean())
print("Test Statistic (t):", t_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("\nDecision: Reject the Null Hypothesis")
    print("Business Conclusion: The average property price in Richmond is significantly different from $1,000,000.")
else:
    print("\nDecision: Fail to Reject the Null Hypothesis")
    print("Business Conclusion: There is no significant evidence that the average property price in Richmond differs from $1,000,000.")
```

```python
# 7
# Independent Two-Sample T-Test (Feature Impact)
# Do properties with car parking sell at a higher average price than properties without car parking, across the entire dataset? Use a 5% significance level and justify:
# • Choice of test
# • Interpretation of p-value
# • Business implications for developers


with_car = df[df['Car'] > 0]['Price'].dropna()
without_car = df[df['Car'] == 0]['Price'].dropna()
t_stat, p_value = stats.ttest_ind(with_car, without_car, equal_var=False)
alpha = 0.05
print("Choice of Test: Independent Two-Sample T-Test (comparing mean prices of two independent groups: properties with car parking vs without car parking)")
print("\nAverage Price (With Car Parking):", with_car.mean())
print("Average Price (Without Car Parking):", without_car.mean())
print("\nTest Statistic (t):", t_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("\nInterpretation of p-value: Since p-value < 0.05, we reject the null hypothesis.")
    print("Conclusion: Properties with car parking sell at a significantly different (higher) average price compared to properties without car parking.")
    print("Business Implication: Developers should prioritize including car parking in property designs as it significantly increases property value.")
else:
    print("\nInterpretation of p-value: Since p-value >= 0.05, we fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference in prices between properties with and without car parking.")
    print("Business Implication: Car parking may not have a strong impact on property price in this dataset.")
```

```python
# 8
# Two-Way ANOVA (Location & Property Type)
# Investigate whether property prices are influenced by:
# • Suburb
# • Type of property
# • Interaction between suburb and property type
# Use Two-Way ANOVA and explain which factors significantly affect price.


import statsmodels.api as sm
from statsmodels.formula.api import ols
data = df[['Price','Suburb','Type']].dropna()
model = ols('Price ~ C(Suburb) + C(Type) + C(Suburb):C(Type)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
alpha = 0.05
print("Two-Way ANOVA Results:\n")
print(anova_table)
print("\nInterpretation:")
for factor in anova_table.index:
    p_value = anova_table.loc[factor, 'PR(>F)']
    if p_value < alpha:
        print(f"{factor} significantly affects property price (p-value = {p_value})")
    else:
        print(f"{factor} does NOT significantly affect property price (p-value = {p_value})")
```

# A hypothesis test comparing prices across two suburbs results in a p-value of 0.032.

#1. What does the p-value indicate?
The p-value of 0.032 means there will be a 3.2% chance of observing the difference in property prices between the two suburbs if the null hypothesis is true. like if both suburbs actually had the same average property price, the probability of getting this difference just by random chance would be very small.

#2. Should the null hypothesis be rejected at α = 0.05?
Yes, because the p-value (0.032) is less than the significance level (0.05), we will reject the null hypothesis.

#3. Business interpretation
This result suggests that there is a statistically significant difference in property prices between the two suburbs. For business stakeholders like real estate developers or property investors, this means that location has a meaningful impact on the property value. Therefore, pricing strategies, investment decisions, and development planning should take these differences between suburbs into account.


```python
# 10
# Industry-Style Hypothesis Validation (Policy Decision)
# A housing policy group believes that properties with more than 2 bathrooms command a premium price.
# Design and execute a statistical test to validate this claim:
# • Identify the correct test
# • State hypotheses
# • Report p-value
# • Give a clear recommendation to policymakers


more_than_2 = df[df['Bathroom'] > 2]['Price'].dropna()
two_or_less = df[df['Bathroom'] <= 2]['Price'].dropna()
t_stat, p_value = stats.ttest_ind(more_than_2, two_or_less, equal_var=False)
alpha = 0.05
print("Test Used: Independent Two-Sample T-Test")
print("\nNull Hypothesis (H0): Properties with more than 2 bathrooms do NOT have a higher average price.")
print("Alternative Hypothesis (H1): Properties with more than 2 bathrooms have a higher average price.")
print("\nAverage Price (>2 Bathrooms):", more_than_2.mean())
print("Average Price (≤2 Bathrooms):", two_or_less.mean())
print("\nTest Statistic:", t_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("\nDecision: Reject the Null Hypothesis")
    print("Recommendation: Properties with more than 2 bathrooms tend to sell at a higher price. Policymakers and developers should encourage larger bathroom configurations as they may increase property value.")
else:
    print("\nDecision: Fail to Reject the Null Hypothesis")
    print("Recommendation: There is no strong statistical evidence that having more than 2 bathrooms significantly increases property price.")
```
