```python
import pandas as pd
df = pd.read_csv('superstore.csv')
df.head()
```

```python
df.info()
```

```python
# Q1: Data Cleaning:
# i. Count and handle Missing values.
# ii. Deal with Duplicate values
# iii. Delete unknown columns if present.
# iv. Check shape, size, and datatypes of the dataset features.

# Q1: Data Cleaning



print("Q1 i: Missing Values Count\n")
print(df.isnull().sum())
df = df.dropna()
print("\nAfter handling missing values, remaining rows:", df.shape[0])
print("\nQ1 ii: Duplicate Values Count\n")
print(df.duplicated().sum())
df = df.drop_duplicates()
print("After removing duplicates, remaining rows:", df.shape[0])
print("\nQ1 iii: Dropping unknown columns if present\n")
if 'ji_lu_shu' in df.columns:
    df = df.drop(columns=['ji_lu_shu'])
    print("Dropped column: ji_lu_shu")
else:
    print("No unknown columns found")
print("\nQ1 iv: Dataset Information\n")
print("Shape (rows, columns):", df.shape)
print("Size (total elements):", df.size)
print("\nData Types:\n")
print(df.dtypes)
```

```python
# Q2: Perform Univariate Analysis across all numerical features,
# i. Which features seem useless in the analysis? Explain why?
# ii. Which features are uniformly distributed or normally distributed?
# iii. Which features are right-skewed/left-skewed? What does this signify?
# iv. Which features have a high number of outliers, and discuss the impact.


import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
num_cols = df.select_dtypes(include=['int64','float64']).columns
print("Numerical Features:\n", num_cols)
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
print("\nSkewness of features:\n")
for col in num_cols:
    print(f"{col}: {skew(df[col].dropna())}")
print("\nOutlier Counts:\n")
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
print("\nQ2 i: Useless Features:")
print("Features like row_id, order_id, customer_id are identifiers and do not provide analytical value for patterns or predictions.")
print("\nQ2 ii: Distribution:")
print("Features with skewness close to 0 are approximately normally distributed. Others are not uniform and show variation in spread.")
print("\nQ2 iii: Skewness Interpretation:")
print("Positive skew (right-skewed) means most values are low with few high extremes (e.g., sales, profit). Negative skew means most values are high with few low extremes.")
print("\nQ2 iv: Outliers Impact:")
print("Features like sales, profit, and shipping_cost typically have high outliers, which can distort averages and affect model performance.")
```

```python
# Q 3: Perform Univariate Analysis across all categorical features.
# i. Which features seem inaccurate and are not useful as categorical “insights” directly?
# ii. What’s the issue with treating Customer? Name as a categorical feature for
# modeling?
# iii. Is the Category distribution balanced or skewed? Give a 1-line reason.
# iv. In Country, does one country dominate strongly? What does that imply about
# geographic bias?
# v. Is the City dataset concentrated in a few cities or spread out?


cat_cols = df.select_dtypes(include=['object']).columns
print("Categorical Features:\n", cat_cols)
for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top Categories in {col}")
    plt.xticks(rotation=45)
    plt.show()
print("\nQ3 i: Inaccurate / Not Useful Features:")
print("Features like customer_id, order_id, product_id, and customer_name are unique identifiers and do not provide meaningful categorical insights.")
print("\nQ3 ii: Issue with Customer Name:")
print("Customer Name has high cardinality (many unique values), which makes it unsuitable for modeling as it does not generalize patterns and increases model complexity.")
print("\nQ3 iii: Category Distribution:")
print("The category distribution is slightly skewed as some categories have more observations than others, indicating imbalance.")
print("\nQ3 iv: Country Dominance:")
print("If one country has significantly more records, it indicates geographic bias, meaning the dataset is not evenly distributed across countries.")
print("\nQ3 v: City Distribution:")
print("The dataset is usually concentrated in a few cities, showing that most business transactions occur in major urban areas rather than being evenly spread.")
```

```python
# Q 4: Perform Bivariant Analysis for numerical-to-numerical features:
# i. Which two features are most strongly correlated?
# ii. Also name features that are negatively correlated.
# iii. If your goal is to understand profit, which are the most useful next bivariate checks?
# Perform them and give clear insights.
# iv. Look for Time Effects Clues. Mention any information you find about the time
# relationship with any feature.



df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
if 'order_date' in df.columns:
    df['order_date'] = pd.to_datetime(df['order_date'])
num_cols = df.select_dtypes(include=['int64','float64']).columns
corr = df[num_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
corr_pairs = corr.unstack().sort_values(ascending=False)
corr_pairs = corr_pairs[corr_pairs != 1]
print("Q4 i: Strongest Positive Correlation:\n", corr_pairs.head(1))
print("\nQ4 ii: Negative Correlations:\n", corr_pairs[corr_pairs < 0].head())
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['sales'], y=df['profit'])
plt.title("Sales vs Profit")
plt.show()
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['discount'], y=df['profit'])
plt.title("Discount vs Profit")
plt.show()
print("\nQ4 iii: Insights:")
print("Sales has a positive relationship with profit.")
print("Discount has a negative relationship with profit.")
if 'order_date' in df.columns:
    df['year'] = df['order_date'].dt.year
    yearly = df.groupby('year')[['sales','profit']].mean()
    yearly.plot(figsize=(8,4))
    plt.title("Yearly Trend of Sales and Profit")
    plt.show()
    print("\nQ4 iv: Time Effect Insight:")
    print("Sales and profit show variation over time, indicating temporal patterns.")
else:
    print("\nQ4 iv: Time Effect Insight:")
    print("Date column not found, unable to analyze time trends.")
```

```python
# Q 5: Perform Bi-variant Analysis for categorical to numerical features.:
# i. Profit by Category: Which category has the highest median profit? Which has the
# lowest? Which category shows the widest spread (largest IQR) in Profit? What does that
# suggest about profit consistency?
# ii. Sales by Category: Which category has the highest median sales? Does it also have
# the highest median profit?
# iii. Profit by Segment: Which segment has the highest median profit? Which segment
# has the most negative/low profit outliers?
# iv. Sales by Segment: Which segment has the highest median sales? Is the profit
# pattern consistent with sales?




df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
plt.figure(figsize=(6,4))
sns.boxplot(x=df['category'], y=df['profit'])
plt.title("Profit by Category")
plt.show()
cat_profit = df.groupby('category')['profit']
median_profit_cat = cat_profit.median()
iqr_cat = cat_profit.quantile(0.75) - cat_profit.quantile(0.25)
print("Q5 i:")
print("Median Profit by Category:\n", median_profit_cat)
print("\nHighest Median Profit Category:", median_profit_cat.idxmax())
print("Lowest Median Profit Category:", median_profit_cat.idxmin())
print("Category with Largest IQR (spread):", iqr_cat.idxmax())
print("Insight: Larger IQR indicates less consistency in profit.")
plt.figure(figsize=(6,4))
sns.boxplot(x=df['category'], y=df['sales'])
plt.title("Sales by Category")
plt.show()
median_sales_cat = df.groupby('category')['sales'].median()
print("\nQ5 ii:")
print("Median Sales by Category:\n", median_sales_cat)
print("Highest Median Sales Category:", median_sales_cat.idxmax())
print("Check if same as highest profit category above.")
plt.figure(figsize=(6,4))
sns.boxplot(x=df['segment'], y=df['profit'])
plt.title("Profit by Segment")
plt.show()
seg_profit = df.groupby('segment')['profit'].median()
print("\nQ5 iii:")
print("Median Profit by Segment:\n", seg_profit)
print("Highest Median Profit Segment:", seg_profit.idxmax())
print("Segment with lowest/negative outliers visible in boxplot.")
plt.figure(figsize=(6,4))
sns.boxplot(x=df['segment'], y=df['sales'])
plt.title("Sales by Segment")
plt.show()
seg_sales = df.groupby('segment')['sales'].median()
print("\nQ5 iv:")
print("Median Sales by Segment:\n", seg_sales)
print("Highest Median Sales Segment:", seg_sales.idxmax())
print("Compare with profit pattern to check consistency.")
```

```python
# Q 6: Perform Bivariate Analysis for Market features against Region, Category, and Country:
# i. Is the Market is not randomly spread across all Regions?
# ii. Which country has negligible office supply orders?
# iii. What are the most useful insights?



df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
market_region = pd.crosstab(df['market'], df['region'])
market_region.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Market vs Region")
plt.xlabel("Market")
plt.ylabel("Count")
plt.show()
print("Q6 i:")
print("Markets are not randomly distributed across regions; certain markets dominate specific regions.")
office = df[df['category'] == 'Office Supplies']
country_office = office['country'].value_counts()
print("\nQ6 ii:")
print(country_office.tail(5))
print("Countries with very low counts have negligible Office Supplies orders.")
market_category = pd.crosstab(df['market'], df['category'])
market_category.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Market vs Category")
plt.show()
print("\nQ6 iii: Insights:")
print("1. Certain markets specialize more in specific product categories.")
print("2. Regional dominance indicates business concentration in certain areas.")
print("3. Some countries have very low order volumes, indicating limited market presence.")
```
